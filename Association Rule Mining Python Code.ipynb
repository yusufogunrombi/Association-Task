# Spain To_list transactions

```python
spain_transactions_ = [spain_transactions.Description[spain_transactions.InvoiceNo == j].to_list() 
                       for j in spain_transactions.InvoiceNo.unique()]
```

```python
rules = list(ap(spain_transactions_, min_support=0.05, min_confidence=0.005, min_length=2))
```
